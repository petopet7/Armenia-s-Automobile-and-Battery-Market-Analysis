## CARS





```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
# Read the CSV file into a DataFrame
cars = pd.read_csv('carss.csv')

# Display the first few rows of the DataFrame and its summary information
cars_info = cars.info()
cars_head = cars.head()

cars_info, cars_head
```

```python
# Sort the DataFrame by date
cars = cars.sort_values('date')

# Plot the sorted data
plt.figure(figsize=(10, 6))
plt.plot(cars['date'], cars['import_amount'], label='Import Amount', marker='o')
plt.plot(cars['date'], cars['export_amount'], label='Export Amount', marker='x')
plt.title('Car Imports and Exports Over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Use MaxNLocator to limit the number of ticks on the x-axis
plt.gca().xaxis.set_major_locator(MaxNLocator(nbins=10))  # Adjust `nbins` to change the number of ticks

plt.tight_layout()
plt.show()
```

```python
# Scatter plot for import_amount vs. export_amount
plt.figure(figsize=(10, 6))
plt.scatter(cars['import_amount'], cars['export_amount'])

# Adding plot labels and title
plt.title('Relationship Between Import and Export Amounts')
plt.xlabel('Import Amount')
plt.ylabel('Export Amount')

# Adding a trend line
z = np.polyfit(cars['import_amount'], cars['export_amount'], 1)
p = np.poly1d(z)
plt.plot(cars['import_amount'], p(cars['import_amount']), "r--")  # Red dashed trend line

plt.grid(True)
plt.show()
```

## BATTERIES

```python
# Load the dataset
data_path = 'batteries.csv'
batteries = pd.read_csv(data_path)

# Convert 'date' column to datetime format for easier plotting
batteries['date'] = pd.to_datetime(batteries['date'])

# Sort the DataFrame by date
batteries.sort_values('date', inplace=True)

# Setting up the plotting area
plt.figure(figsize=(14, 10))

# Plotting Export and Import quantities over time
plt.subplot(2, 1, 1)
plt.plot(
    batteries['date'], batteries['export_in_tonn'],
    label='Export Quantity (Tonn)', marker='o'
)
plt.plot(
    batteries['date'], batteries['import_in_tonn'],
    label='Import Quantity (Tonn)', marker='x'
)
plt.title('Export vs Import Quantities over Time')
plt.xlabel('Date')
plt.ylabel('Quantity in Tons')
plt.legend()

# Plotting Export and Import values over time
plt.subplot(2, 1, 2)
plt.plot(
    batteries['date'], batteries['export_in_1000_'],
    label='Export Value (1000$)', marker='o'
)
plt.plot(
    batteries['date'], batteries['import_in_1000_'],
    label='Import Value (1000$)', marker='x'
)
plt.title('Export vs Import Values over Time')
plt.xlabel('Date')
plt.ylabel('Value in 1000$')
plt.legend()

plt.tight_layout()
plt.show()



```

```python
# Reload the dataset
data_path = 'batteries.csv'
batteries_data = pd.read_csv(data_path)
batteries_data['date'] = pd.to_datetime(batteries_data['date'])

# Ensure the data is sorted by date in ascending order
batteries_data_sorted = batteries_data.sort_values(by='date')

# Setting up the plotting parameters again after sorting
plt.figure(figsize=(16, 14))  # Adjusted figure size for clarity
bar_width = 0.35
index_sorted = np.arange(len(batteries_data_sorted['date']))

# Calculate the step for displaying labels to reduce clutter, showing every fourth label
label_step = 4
date_labels = [
    date.strftime('%Y-%m') if idx % label_step == 0 else ''
    for idx, date in enumerate(batteries_data_sorted['date'])
]

# Bar chart for comparing monthly export and import quantities with fewer x-axis labels
plt.subplot(2, 1, 1)
plt.bar(
    index_sorted, batteries_data_sorted['export_in_tonn'],
    bar_width, label='Export Quantity (Tonn)'
)
plt.bar(
    index_sorted + bar_width, batteries_data_sorted['import_in_tonn'],
    bar_width, label='Import Quantity (Tonn)'
)

plt.xlabel('Date')
plt.ylabel('Quantity in Tons')
plt.title('Monthly Export vs Import Quantities')
plt.xticks(index_sorted + bar_width / 2, date_labels, rotation=90)  # Apply adjusted labels
plt.legend()

# Bar chart for comparing monthly export and import values with fewer x-axis labels
plt.subplot(2, 1, 2)
plt.bar(
    index_sorted, batteries_data_sorted['export_in_1000_'],
    bar_width, label='Export Value (1000$)'
)
plt.bar(
    index_sorted + bar_width, batteries_data_sorted['import_in_1000_'],
    bar_width, label='Import Value (1000$)'
)

plt.xlabel('Date')
plt.ylabel('Value in 1000$')
plt.title('Monthly Export vs Import Values')
plt.xticks(index_sorted + bar_width / 2, date_labels, rotation=90)  # Apply adjusted labels
plt.legend()

plt.tight_layout()
plt.show()

```

# Car Listings


```python
# Load the CSV file to get an overview of the data
data_path = 'carsbat.csv'
cars_data = pd.read_csv(data_path)

# Display the first few rows of the dataset to understand its structure
cars_data.head()
```

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

fig, axs = plt.subplots(3, 2, figsize=(14, 18))

# Distribution of Cars by Year of Manufacture
sns.histplot(data=cars_data, x="Year", kde=True, color="skyblue", ax=axs[0, 0])
axs[0, 0].set_title('Distribution of Cars by Year of Manufacture')

# Battery Size Distribution
sns.countplot(
    data=cars_data, y="battery",
    order=cars_data['battery'].value_counts().index, ax=axs[0, 1]
)
axs[0, 1].set_title('Battery Size Distribution')

# Mileage Distribution
sns.histplot(
    data=cars_data, x="mileage_in_km",
    bins=30, color="olive", ax=axs[1, 0], kde=True
)
axs[1, 0].set_title('Mileage Distribution')
axs[1, 0].set(xscale="log")  # Using a log scale due to wide range of values

# Price Distribution
sns.histplot(
    data=cars_data, x="price_in_usd",
    bins=30, color="gold", ax=axs[1, 1], kde=True
)
axs[1, 1].set_title('Price Distribution')

# Number of Cars by Brand
sns.countplot(
    data=cars_data, y="Car_Brand",
    order=cars_data['Car_Brand'].value_counts().index, ax=axs[2, 0]
)
axs[2, 0].set_title('Number of Cars by Brand')

# Adjust layout and remove the empty subplot
plt.tight_layout()
plt.delaxes(axs[2][1])  # Remove the empty subplot
plt.show()

```

```python
data = cars_data
# Convert 'AGM' to a binary column for visualization purposes
data['AGM'] = data['AGM'].fillna('No').replace('Y', 'Yes')

# Set up the visualization style
sns.set_style('whitegrid')

# Create the plot
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Year', hue='AGM', palette='coolwarm')

# Add labels and a title
plt.xlabel('Car Manufacturing Year')
plt.ylabel('Number of Cars')
plt.title('Distribution of Car Manufacturing Years with AGM Battery Information')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Load the data
file_path = 'carsbat.csv'
cars_data = pd.read_csv(file_path)

# Clean the data
# Drop rows where 'battery' is NaN
cars_data.dropna(subset=['battery'], inplace=True)

# Extract numeric values from 'battery' strings
cars_data['battery_size'] = cars_data['battery'].str.extract('(\d+)').astype(int)

# Normalize 'AGM' column: treat any non-null as 'AGM', nulls as 'Non-AGM'
cars_data['AGM'] = cars_data['AGM'].apply(lambda x: 'AGM' if pd.notna(x) else 'Non-AGM')

# Aggregate data
aggregated_data = cars_data.groupby(['battery_size', 'AGM']).size().unstack(fill_value=0)

# Calculate the total count and percentage for overall battery sizes
total_batteries = cars_data['battery_size'].value_counts().sum()
aggregated_data['Total'] = aggregated_data.sum(axis=1)
plot_data = pd.DataFrame(aggregated_data['Total'])
plot_data['Percentage'] = (plot_data['Total'] / total_batteries) * 100

# Plotting
fig, ax = plt.subplots(figsize=(10, 8))
aggregated_data[['Non-AGM', 'AGM']].plot(kind='barh', stacked=True, color=['#d62728', '#1f77b4'], ax=ax)
ax.set_xlabel('Count')
ax.set_ylabel('Battery Size (Ah)')
ax.set_title('Battery Size Distribution with AGM and Non-AGM')

# Add overall percentages on each bar
for i, (index, row) in enumerate(plot_data.iterrows()):
    ax.text(row['Total'] + 5, i, f'{round(row["Percentage"], 2)}%', ha='left', va='center', color='black')

plt.show()


```
